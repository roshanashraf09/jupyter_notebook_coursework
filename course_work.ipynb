Exercise 1 - Prime Number Checker

```python
num = int(input("Enter a number: "))

if num < 2:
    print("This is not a prime number")
else:
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1):  # only need to check up to √num
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print("This is a prime number")
    else:
        print("This is not a prime number")
```
