Exercise 1 - Prime Number Checker

```python
num = int(input("Enter a number: "))

if num < 2:
    print("This is not a prime number")
else:
    is_prime = True
    for i in range(2, int(num ** 0.5) + 1): 
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print("This is a prime number")
    else:
        print("This is not a prime number")
```

Exercise 5 - Min and Max of an iterable

```python
def min_max(collection):

    return min(collection), max(collection)

numbers = [4, 7, 1, 9, 3, 15, 2]

low, high = min_max(numbers)  

print("Min value is:", low)
print("Max value is:", high)
```

Exercise 6 - Guess the Number Game

```python
import random

secret = random.randint(1, 10)  
while True:
    guess = int(input("Guess the number (1-10): "))

    if guess > secret:
        print("Too high! Try again.")
    elif guess < secret:
        print("Too low! Try again.")
    else:
        print(f"Correct! The number was {secret}. You win!")
        break 
```

Exercise 8 - Mathematical Series

```python
import math

def calculate_series(x, n):
   
    total = 0

    for i in range(1, n + 1):
        numerator   = x * (5 ** i) ** 2          # x * (5^i)^2
        denominator = i + math.factorial(i + 1)  # i + (i+1)!
        total += numerator / denominator

    return total

# Get user inputs
x = float(input("Enter the value of x: "))
n = int(input("Enter the number of terms (N): "))

result = calculate_series(x, n)
print(f"Sum of series with x={x} and N={n} terms: {result:.6f}")
```

Exercise 9 - Image Mirror Reflections

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread("./images/doraemon.jpg")  
img_array = np.array(img)            

horizontal_flip = img_array[:, ::-1, :]  
vertical_flip   = img_array[::-1, :, :]  

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(img_array)
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(horizontal_flip)
axes[1].set_title("Horizontal Mirror")
axes[1].axis("off")

axes[2].imshow(vertical_flip)
axes[2].set_title("Vertical Mirror")
axes[2].axis("off")

plt.tight_layout()
plt.show()
```

Exercise 10 - Temperature Data Analysis with NumPy

```python
import numpy as np

temps_celsius = np.array([
    22, 24, 19, 21, 23, 25, 26, 27, 24, 23,
    20, 22, 21, 19, 18, 22, 23, 24, 26, 28,
    30, 27, 25, 26, 24, 22, 21, 20, 19, 23
])

print("=== Basic Statistics ===")
print(f"Minimum temperature : {np.min(temps_celsius)}°C")
print(f"Maximum temperature : {np.max(temps_celsius)}°C")
print(f"Average temperature : {np.mean(temps_celsius):.2f}°C")

temps_fahrenheit = (temps_celsius * 9 / 5) + 32
print("\n=== Fahrenheit Conversion ===")
print("Temperatures in Fahrenheit:", temps_fahrenheit)

above_25 = temps_celsius[temps_celsius > 25]
days_above_25 = np.where(temps_celsius > 25)[0] + 1 
print("\n=== Days Above 25°C ===")
print("Day numbers:", days_above_25)
print("Temperatures:", above_25)

days_below_20 = np.sum(temps_celsius < 20) 
print("\n=== Days Below 20°C ===")
print(f"Number of days below 20°C: {days_below_20}")

temps_sorted = np.sort(temps_celsius)
print("\n=== Sorted Temperatures ===")
print(temps_sorted)

temps_2d = temps_celsius.reshape(5, 6)
print("\n=== Reshaped to 5 Weeks × 6 Days ===")
print(temps_2d)
```

Exercise 11 - Pandas DataFrame with Random Numbers
