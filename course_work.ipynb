## 1. Prime number checker
I wanted to keep this one simple. The program asks for a number and checks whether it has any factors other than 1 and itself.


```python
number = int(input("Enter a whole number: "))

if number < 2:
    print("Not a prime number")
else:
    prime = True
    for divisor in range(2, int(number ** 0.5) + 1):
        if number % divisor == 0:
            prime = False
            break

    if prime:
        print("Prime number")
    else:
        print("Not a prime number")
```

## 2. Fibonacci sequence
 This prints the first few terms of the Fibonacci sequence by updating two values each time through the loop.


```python
terms = int(input("How many Fibonacci terms do you want? "))

first = 0
second = 1
for _ in range(terms):
    print(first, end=" ")
    first, second = second, first + second

print()
```

## 3. Star pattern
For this exercise I printed a pattern that gets smaller and then larger again. I hard-coded the size to 5 just to keep it quick.


```python
rows = 5

for count in range(rows, 0, -1):
    print("* " * count)

for count in range(2, rows + 1):
    print("* " * count)
```

## 4. Reverse a string
This one uses slicing to reverse whatever text the user types in.


```python
text = input("Enter some text: ")
print("Reversed:", text[::-1])
```

## 5. Minimum and maximum values
I used Python's built-in `min()` and `max()` functions here and wrapped them in a small function.


```python
def find_min_max(values):
    return min(values), max(values)

numbers = [4, 7, 1, 9, 3, 15, 2]
smallest, largest = find_min_max(numbers)

print("Min value:", smallest)
print("Max value:", largest)
```

## 6. Guess the number game
This generates a random number from 1 to 10 and keeps asking for guesses until the user gets it right.


```python
import random

secret_number = random.randint(1, 10)

while True:
    guess = int(input("Guess a number between 1 and 10: "))

    if guess > secret_number:
        print("Too high, try again.")
    elif guess < secret_number:
        print("Too low, try again.")
    else:
        print("You got it!")
        break
```

## 7. Password checker
The last exercise is a basic password check using a loop so the user keeps trying until the correct password is entered.


```python
correct_password = "password"

while True:
    user_password = input("Enter password: ")

    if user_password == correct_password:
        print("Access granted")
        break

    print("Wrong password, try again.")
```
